```python
import numpy as np
import matplotlib.pyplot as plt

# General likelihood for k heads, n-k tails
def likelihood(theta, k, n):
    return (theta*k) * ((1-theta)*(n-k))

def log_likelihood(theta, k, n):
    return np.log(likelihood(theta, k, n))

def neg_log_likelihood(theta, k, n):
    return -np.log(likelihood(theta, k, n))

# Function to plot all three
def plot_case(k, n):
    theta_range = np.linspace(0.001, 0.999, 500)  # avoid log(0)
    
    L_vals = likelihood(theta_range, k, n)
    logL_vals = log_likelihood(theta_range, k, n)
    neglogL_vals = neg_log_likelihood(theta_range, k, n)
    
    # Find MLE
    mle_theta = k/n
    mle_value = likelihood(mle_theta, k, n)
    
    # Plot
    plt.figure(figsize=(15,4))
    
    # Likelihood
    plt.subplot(1,3,1)
    plt.plot(theta_range, L_vals, label="L(θ)")
    plt.axvline(mle_theta, color="green", linestyle="--", label=f"MLE = {mle_theta:.2f}")
    plt.title(f"Likelihood (k={k}, n={n})")
    plt.xlabel("θ")
    plt.ylabel("L(θ)")
    plt.legend()
    
    # Log-Likelihood
    plt.subplot(1,3,2)
    plt.plot(theta_range, logL_vals, label="log L(θ)")
    plt.axvline(mle_theta, color="green", linestyle="--")
    plt.title("Log-Likelihood")
    plt.xlabel("θ")
    plt.ylabel("log L(θ)")
    plt.legend()
    
    # Negative Log-Likelihood
    plt.subplot(1,3,3)
    plt.plot(theta_range, neglogL_vals, label="-log L(θ)")
    plt.axvline(mle_theta, color="green", linestyle="--")
    plt.title("Negative Log-Likelihood")
    plt.xlabel("θ")
    plt.ylabel("-log L(θ)")
    plt.legend()
    
    plt.tight_layout()
    plt.show()


# Example runs
plot_case(90, 100)   # 90 heads, 10 tails
plot_case(80, 100)  # 80 heads, 20 tails
plot_case(75, 10)  # 75 heads, 25 tails
plot_case(60, 100) # 60 heads, 40 tails
plot_case(50, 100)  # 50 heads, 50 tails
plot_case(45, 100)  # 45 heads, 55 tails
plot_case(40, 100)  # 40 heads, 60 tails
plot_case(30, 100)  # 30 heads, 70 tails
plot_case(25, 100)  # 25 heads, 75 tails
plot_case(10,100)   # 10 heads, 90 tails
```
